Create a 1D tensor from the Python list [1, 2, 3, 4, 5] and print its shape. Then, reshape it to a 2D tensor of shape (5, 1) and print the new tensor.

```python
a = torch.tensor([1, 2, 3, 4, 5])
print(a)
print(a.shape)

a_reshaped = a.view(5, 1)
print( a_reshaped)
```

Write a Python program that creates a tensor of shape (3, 3) filled with random values between 0 and 1 using PyTorch. Print the tensor.

```python
b = torch.rand(3, 3)
print(b)
```

Create a tensor filled with zeros of shape (2, 4) using PyTorch. Then, slice this tensor to get the first row and print it.

```python
a = torch.zeros(2, 4)
print(a)
print(a[0])
```

Using PyTorch, create two 2D tensors A of shape (3, 2) and B of shape (2, 3). Perform matrix multiplication between A and B and print the result.

```python
a = torch.rand(3, 2)
b = torch.rand(2, 3)
print(a)
print(b)
print(torch.matmul(a, b))
```

Write a Python function using PyTorch to compute the element-wise addition and multiplication of two tensors. Create two random tensors of the same shape and perform these operations.

```python
a = torch.rand(2, 2) * 5
b = torch.rand(2, 2) + 11
c = a * b
d = a + b

print(c)
print(d)
```

Write a Python program that creates a 2D tensor of shape (3, 3) filled with ones. Then, use PyTorch to concatenate this tensor along the columns with another tensor of shape (3, 2) filled with random values.

```python
ones = torch.ones(3, 3)
rand = torch.rand(3, 2)
c = torch.concat((ones, rand), dim=1)
print(c)
```

Create a tensor of shape (4, 4) filled with random integers between 1 and 10 using PyTorch. Then, reshape the tensor to (2, 8) and print both the original and reshaped tensors.

```python
a = torch.randint(1, 11, (4, 4))
reshaped = a.view(2, 8)
print(a)
print(reshaped)
```
